```python
import numpy as np
```

```python
emp_id = np.array([101, 102, 103, 104, 105, 106, 107, 108, 109, 110])

salary = np.array([45000, 55000, 70000, 35000, 80000,
                   60000, 50000, 90000, 65000, 40000])

department = np.array([
    "HR", "IT", "Finance", "HR", "IT",
    "Finance", "HR", "IT", "Finance", "HR"
])
```

```python
print("Employee IDs:", emp_id)
print("Salaries:", salary)
print("Departments:", department)
```

```python
print("Total Employees", len(emp_id))
```

## Salary Statistics Analysis

In this section, we calculate:
- Average Salary
- Highest Salary
- Lowest Salary

These metrics help understand the salary distribution within the organization.

```python
print("Average Salary:", np.mean(salary))
print("Highest Salary:", np.max(salary))
print("Lowest Salary:", np.min(salary))
```

## Employees Earning Above Average Salary

Employees whose salaries exceed the company average are identified below.

```python
avg_salary = np.mean(salary)

for i in range(len(salary)):
    if salary[i] > avg_salary:
        print(emp_id[i], salary[i])
```

```python
for dept in np.unique(department):
    dept_salary = salary[department == dept]
    print(dept, "Average Salary:", np.mean(dept_salary))
```

```python
bonus_salary = salary * 1.10
print(bonus_salary)
```

## Business Insights

- The average employee salary was calculated successfully.
- IT department has competitive salary levels.
- Several employees earn above the company average.
- A 10% bonus impact was analyzed for all employees.
